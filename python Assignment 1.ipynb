```python
#the sentence to uppercase.
strM = "Python beginner tutorial"
print(f"the sentence to uppercase:", strM.upper())
```

```python
#the sentence to lowercase.
print(f"the sentence to lowercase:", strM.lower())
```

```python
#Use Capitalize and return the sentence to the original input form.
print(f"the Capitalize sentence:",strM.title())
```

```python
#Count the total number of occurrences of character ‘t’ in the string.
print(f"Counting the string't'is: ", strM.count('t'))
```

```python
#Replace all occurrences of “Python” with “Machine Learning” in the input string
print(f"Replace the python to Machine learning: ", strM.replace("Python", "Machine learning"))
```
